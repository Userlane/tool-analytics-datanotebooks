```python
# package this as a function

def feature_engineer_dataset(data_path = "data/UnAppIt1", propertyId = "31610"):
    df = pd.read_csv(f"{data_path}.csv", index_col = 0)
    df = df[df.propertyId == propertyId]
    df1 = df.groupby(["subjectValue", "action", "appSessionId"]).count().reset_index().loc[:,
               ["subjectValue", "action", "appSessionId","clientTime"]].rename(columns = {"clientTime":"count"})
    df1 = pd.merge(df1, df.loc[:,["appSessionId", "TimeDiff"]].drop_duplicates(), how = "left", on = "appSessionId")
    df2 = pd.pivot_table(df1, values="count", index = "appSessionId", columns = "action").reset_index().fillna(0)
    df3 = pd.merge(df1.loc[:,["subjectValue", "appSessionId", "TimeDiff"]], df2, how = "left", on = "appSessionId")
    
    df3.to_csv(f"{data_path}_{propertyId}.csv")
```

```python
feature_engineer_dataset(propertyId = "31610")
```
